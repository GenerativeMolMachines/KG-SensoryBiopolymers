# Parsing sequences and smiles

```python
import time
import requests
from tqdm import tqdm
```

```python
def get_smiles_from_pubchem(compound_name):
    try:

        url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/{compound_name}/property/CanonicalSMILES/JSON"
        response = requests.get(url)

        if response.status_code == 200:
            data = response.json()
            smiles = data['PropertyTable']['Properties'][0]['CanonicalSMILES']

            time.sleep(1)
            return smiles
        else:
            print(f"Fail for {compound_name}. Response code: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error during processing {compound_name}: {e}")
        return None

unique_compounds = df_chem['name'].unique()

smiles_dict = {}
for compound_name in tqdm(unique_compounds, desc="Fetching SMILES"):
    smiles = get_smiles_from_pubchem(compound_name)
    smiles_dict[compound_name] = smiles

df_chem['content'] = df_chem['name'].map(smiles_dict)

print(df_chem.info())
print(df_chem.head())
```

```python
df_chem.reset_index(drop=True, inplace=True)
df_chem.to_csv('chemical_conditions.csv', index=False)

unique_mol = pd.DataFrame()
unique_mol[['name', 'hmdb_id', 'content']] = df_chem[['name', 'hmdb_id', 'content']]
unique_mol = unique_mol.drop_duplicates()
unique_mol.reset_index(drop=True, inplace=True)
unique_mol.to_csv('unique_chemical_biomarkers.csv', index=False)
```

```python
def fetch_uniprot_sequence(uniprot_id):
    try:
        url = f"https://www.uniprot.org/uniprot/{uniprot_id}.fasta"
        response = requests.get(url)

        if response.status_code == 200:
            fasta_data = response.text
            sequence = ''.join(fasta_data.split('\n')[1:])

            time.sleep(1)
            return sequence
        else:
            print(f"Fail for {uniprot_id}. Response code: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error during processing {uniprot_id}: {e}")
        return None


unique_uniprot_ids = df_prot['uniprot_id'].unique()

sequence_dict = {}
for uniprot_id in tqdm(unique_uniprot_ids, desc='Fetching sequences'):
    sequence = fetch_uniprot_sequence(uniprot_id)
    sequence_dict[uniprot_id] = sequence

df_prot['content'] = df_prot['uniprot_id'].map(sequence_dict)
```

```python
df_prot.to_csv('protein_condition.csv', index=False)
```

```python
unique_proteins = pd.DataFrame()
unique_proteins[['name', 'uniprot_id', 'content']] = df_prot[['name', 'uniprot_id', 'content']]
unique_proteins = unique_proteins.drop_duplicates()
unique_proteins.reset_index(drop=True, inplace=True)
unique_proteins.to_csv('unique_protein_biomarkers.csv', index=False)
```

```python
unique_conditions_prot = pd.DataFrame()
unique_conditions_prot['conditions'] = df_prot['conditions']
unique_conditions_prot = unique_conditions_prot.drop_duplicates()
unique_conditions_prot.reset_index(drop=True, inplace=True)

unique_conditions_mol = pd.DataFrame()
unique_conditions_mol['conditions'] = df_chem['conditions']
unique_conditions_mol = unique_conditions_mol.drop_duplicates()
unique_conditions_mol.reset_index(drop=True, inplace=True)

unique_conditions = pd.concat([unique_conditions_mol, unique_conditions_prot], axis=0)
unique_conditions = unique_conditions.drop_duplicates()
unique_conditions.reset_index(drop=True, inplace=True)
unique_conditions.to_csv('unique_conditions.csv', index=False)
```
