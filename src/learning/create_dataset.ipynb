```python
antibodies = pd.read_csv('drive/MyDrive/нирси/antibodies.csv')
antibody_inter = pd.read_csv('drive/MyDrive/нирси/antibody_interactions.csv')
aptamers = pd.read_csv('drive/MyDrive/нирси/aptamers.csv')
aptamer_inter = pd.read_csv('drive/MyDrive/нирси/aptamers_interactions.csv')
molecules_from_antibodies = pd.read_csv('drive/MyDrive/нирси/molecules_from_antibodies.csv')
molecules_from_aptamer = pd.read_csv('drive/MyDrive/нирси/molecules_from_aptamers.csv')
```

```python
triples = []

for _, row in antibodies.iterrows():
    name = row["name"]
    type_val = row["type"]
    subtype_val = row["subtype"]

    triples.append((name, "class_of", type_val))
    if pd.notnull(subtype_val) and subtype_val != "":
        triples.append((name, "subclass_of", subtype_val))

for _, row in antibody_inter.iterrows():
    antibody_name = row["antibody_name"]
    ligand_name = row["ligand_name"]
    if pd.notnull(ligand_name) and ligand_name != "":
        triples.append((antibody_name, "interacts_with", ligand_name))

for _, row in aptamers.iterrows():
    name = row["name"]
    type_val = row["type"]
    subtype_val = row["subtype"]

    triples.append((name, "class_of", type_val))
    if pd.notnull(subtype_val) and subtype_val != "":
        triples.append((name, "subclass_of", subtype_val))

for _, row in aptamer_inter.iterrows():
    aptamer_name = row["aptamer_name"]
    target_name = row["target_name"]
    if pd.notnull(target_name) and target_name != "":
        triples.append((aptamer_name, "interacts_with", target_name))

for _, row in molecules_from_antibodies.iterrows():
    molecule_name = row["name"]
    type_val = row["type"]
    if pd.notnull(molecule_name) and pd.notnull(type_val):
        triples.append((molecule_name, "class_of", type_val))

for _, row in molecules_from_aptamer.iterrows():
    molecule_name = row["name"]
    type_val = row["type"]
    if pd.notnull(molecule_name) and pd.notnull(type_val):
        triples.append((molecule_name, "class_of", type_val))

triples.append(("Aptamers", "subclass_of", "Nucleic Acids"))
triples.append(("DNA", "subclass_of", "Nucleic Acids"))
triples.append(("RNA", "subclass_of", "Nucleic Acids"))

triples_df = pd.DataFrame(triples, columns=["subject", "predicate", "object"])
```
